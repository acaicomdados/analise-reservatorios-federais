```python
import pandas as pd
```

```python
estacao = pd.read_csv('dado estação.csv')
estacao.head()
```

```python
estacao = estacao.iloc[:,0:12]
estacao.head()
```

```python
estacao['uf'].unique()
```

```python
estacao = estacao.query('uf!="EQ" and uf!="PU" and uf!="GF" and uf!="AR" and uf!="PG" and uf!="UR" and uf!="GU" and uf!="SU" and uf!="CO"')
estacao.head()
```

```python
estacao['uf'].unique()

estacao = estacao.query('uf!="BO"')
estacao.head()
```

```python
import seaborn as sns
sns.histplot(data=estacao,x='alt',kde=True)
```

```python
import plotly.express as px
fig = px.scatter_mapbox(estacao,lat='lat',lon='lon',hover_name='name')
fig.update_layout(mapbox_style="open-street-map")
fig.show()
```

```python
estacao.to_csv('estacao pluviometrica.csv')
```
