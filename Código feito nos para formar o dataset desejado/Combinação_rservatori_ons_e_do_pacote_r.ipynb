```python
import basedosdados as bd
import pandas as pd
```

```python
gera_ear = bd.read_table(dataset_id='br_ons_avaliacao_operacao',
table_id='reservatorio',
billing_project_id="1021231706388")
gera_ear.head()
```

```python
gera_ear.drop(['id_subsistema'],axis=1,inplace=True)
gera_ear.head()
```

```python
gera_ear.iloc[:,0:13]
```

```python
gera_ear.iloc[:,13:23]
```

```python
url = 'https://raw.githubusercontent.com/acaicomdados/analise-reservatorios-federais/main/Datasets%20/lista_reservatorios.csv'
lista_reservatorios = pd.read_csv(url)
lista_reservatorios.head()
```

```python
lista_reservatorios.drop(['municipio','codigo_municipio_ibge','Unnamed: 0'],axis=1,inplace=True)
lista_reservatorios.head()
```

```python
lista_reservatorios.drop(['res_latitude','res_longitude','estado_nome'],axis=1,inplace=True)
lista_reservatorios.head()
```

```python
lista_reservatorios.shape
```

```python
lista_reservatorios.drop_duplicates(inplace=True)
lista_reservatorios.shape
```

```python
lista_reservatorios.info()
```

```python
lista_reservatorios.drop(['rio','sub_bacia','bacia'],axis=1,inplace=True)
lista_reservatorios.head()
```

```python
reservatorio = gera_ear.merge(lista_reservatorios,on='reservatorio')
reservatorio.head()
```

```python
reservatorio.iloc[:,0:13].head()
```

```python
reservatorio.iloc[:,13:26].head()
```

```python
reservatorio.to_csv('reservatorio_geral_codigo_aneel.csv')
```
