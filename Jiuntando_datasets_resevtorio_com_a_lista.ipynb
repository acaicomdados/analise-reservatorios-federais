```python
import numpy as np
import pandas as pd
```

```python
lista_reservatorio = pd.read_csv('lista_reservatorios.csv')
lista_reservatorio.drop('Unnamed: 0',axis=1,inplace=True)
lista_reservatorio.head()
```

```python
lista_reservatorio.rename(columns={'codigo':'id_reservatorio'},inplace=True)
lista_reservatorio.head()
```

```python
lista_reservatorio.info()
```

```python
reservatorio = pd.read_csv('/content/drive/MyDrive/Datasets/Reservatorio Federal.csv')
reservatorio.drop('Unnamed: 0',axis=1,inplace=True)
reservatorio.head()
```

```python
reservatorio.info()
```

```python
reservatorio = reservatorio.merge(lista_reservatorio,on='id_reservatorio')
reservatorio.head()
```

```python
reservatorio.info()
```

```python
reservatorio.drop(['reservatorio','codigo_municipio_ibge','estado_sigla'],
                  axis=1,inplace=True)
reservatorio.head()
```

```python
reservatorio['nome_reservatorio'] = reservatorio['nome_reservatorio'].str.capitalize()
reservatorio['municipio'] = reservatorio['municipio'].str.capitalize()
reservatorio['rio'] = reservatorio['rio'].str.capitalize()
reservatorio['sub_bacia'] = reservatorio['sub_bacia'].str.capitalize()
reservatorio['bacia'] = reservatorio['bacia'].str.capitalize()

reservatorio.head()
```

```python
reservatorio.drop('sistema',axis=1,inplace=True)
reservatorio.head()
```

```python
reservatorio.to_csv('Reservatorio_com_outras_info.csv')
```
