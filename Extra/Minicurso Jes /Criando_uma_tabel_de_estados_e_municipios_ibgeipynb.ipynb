Neste notebbok, com base no github do leo germani (sendo que o repositório pode ser acessado [aqui](https://github.com/leogermani/estados-e-municipios-ibge)) vamos gerar um tabela com o código do municipio do ibge, o nome da cidade e o Estado onde se localiza o muncipio

```python
import pandas as pd
```

```python
url = 'https://raw.githubusercontent.com/leogermani/estados-e-municipios-ibge/master/estados.csv'

estado = pd.read_csv(url)
estado.head()
```

```python
estado.rename(columns={'COD': 'COD UF'}, inplace=True)
estado = estado[['COD UF', 'SIGLA']]
estado.head()
```

```python
url2 = 'https://raw.githubusercontent.com/leogermani/estados-e-municipios-ibge/master/municipios.csv'

municipio = pd.read_csv(url2)
municipio.head()
```

```python
estado_e_municipio = pd.merge(estado, municipio, on='COD UF')
estado_e_municipio.head()
```

```python
estado_e_municipio.rename(columns={'SIGLA': 'Sigla_estado','NOME':'Nome do municipio'}, inplace=True)
estado_e_municipio.head()
```

```python
estado_e_municipio.to_csv('estado_e_municipio.csv')
```
